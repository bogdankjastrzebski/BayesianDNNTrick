```julia
using Plots
using Zygote
theme(:default, background_color=:transparent)
```

```julia
f(x, α) = 15 * (x-α)^2 * (x+α)^2 / (16*α^5)
```

```julia
X = [-1,1]
```

```julia
L(α) = prod(f(x, α) for x in X)
```

```julia
plot(L, 1, 20, label="L") # x ∈ [-α, α] i x₁ = 1 daje nam, że α ≥ 1, stąd lewa krawędź wykresu jest przy 1.
plot!([√5, √5], [0, 0.075], color="red", label="α̂ₘₗₑ") # nasze rozwiązanie
```

```julia
plot(log ∘ L, 1, 20, label="logL")
plot!([√5, √5], [-25, 0.0], color="red", label="α̂ₘₗₑ") # nasze rozwiązanie
```

```julia
plot((log ∘ L)', 1, 20, label="∇logL")
plot!([√5, √5], [-1, 10.0], color="red", label="α̂ₘₗₑ") # nasze rozwiązanie
ylims!(-1,10)
```
