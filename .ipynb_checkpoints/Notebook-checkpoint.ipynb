### Dlaczego rozkład normalny nazywa się normalny

```julia
plot(x -> pdf(Beta(2,3), x))
plot!(x -> pdf(Exponential(1), x))
plot!(x -> pdf(Normal(0,0.2), x))
```

```julia
function sample_sum(d, n=1000)
    r = rand(d, n)
    return sum(r)
end
```

```julia
histogram([sample_sum(Bernoulli(0.1),10000) for _ in 1:10000])
```

```julia
X = randn(100)
Y =  π * X + 0.5randn(100)
# Y = 3X.^2 + 2X .+ 1 + 0.1randn(100)
scatter(X, Y)
```

```julia
L(a) = sum((a * X - Y).^2)
```

```julia
plot(L)
```

```julia
x = 0.0
α = 0.0001
xs = []
for _ in 1:1000
    x -= α*L'(x)
    push!(xs, x)
end
```

```julia
plot(L)
plot!(xs, L.(xs))
```
